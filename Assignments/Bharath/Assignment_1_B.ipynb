```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
pd.set_option("display.max_column",None)
pd.set_option("display.max_row",None)
```

```python
data=pd.read_csv("/content/drive/MyDrive/IBM/data/House Price India.csv")
data.head()
```

## **Univariate Analysis**

```python
data.dropna(inplace=True)
```

```python
fig,axes = plt.subplots(2,2,figsize=(16,8))
sns.countplot(x="number of bedrooms",data=data,ax=axes[0,0])
sns.countplot(x="number of bathrooms",data=data,ax=axes[0,1])
sns.countplot(x="waterfront present",data=data,ax=axes[1,0])
sns.countplot(x="number of floors",data=data,ax=axes[1,1])


```

```python
fig_c,axes_c = plt.subplots(2,2,figsize=(16,8))
sns.countplot(x="condition of the house",data=data,ax=axes_c[0,0])
sns.countplot(x="Number of schools nearby",data=data,ax=axes_c[0,1])
sns.countplot(x="number of views",data=data,ax=axes_c[1,0])
sns.countplot(x="grade of the house",data=data,ax=axes_c[1,1])
```

```python
fig,axes=plt.subplots(2,2,figsize=(16,8))
sns.histplot(x="grade of the house",data=data,ax=axes[0,0])
sns.histplot(x="Price",data=data,ax=axes[0,1])
sns.histplot(x="number of views",data=data,ax=axes[1,0])
sns.histplot(x="condition of the house",data=data,ax=axes[1,1])
```

```python
fig,axes=plt.subplots(2,2,figsize=(16,8))
sns.boxplot(x="grade of the house",data=data,ax=axes[0,0])
sns.boxplot(x="Price",data=data,ax=axes[0,1])
sns.boxplot(x="number of views",data=data,ax=axes[1,0])
sns.boxplot(x="condition of the house",data=data,ax=axes[1,1])
```

## Bivariate analysis

### Category/Numerical

```python
fig,axes=plt.subplots(3,2,figsize=(15,10))
sns.barplot(x="grade of the house",y="Price",data=data,ax=axes[0,0])
sns.barplot(x="condition of the house",y="Price",data=data,ax=axes[0,1])
sns.barplot(x="number of bedrooms",y="Price",data=data,ax=axes[1,0])
sns.barplot(x="number of bathrooms",y="Price",data=data,ax=axes[1,1])
sns.barplot(x="waterfront present",y="Price",data=data,ax=axes[2,0])
sns.barplot(x="number of views",y="Price",data=data,ax=axes[2,1])
```

### Numerical/numerical

```python
fig,axes=plt.subplots(3,2,figsize=(15,10))
sns.scatterplot(x="grade of the house",y="Price",data=data,ax=axes[0,0])
sns.scatterplot(x="condition of the house",y="Price",data=data,ax=axes[0,1])
sns.scatterplot(x="number of bedrooms",y="Price",data=data,ax=axes[1,0])
sns.scatterplot(x="number of bathrooms",y="Price",data=data,ax=axes[1,1])
sns.scatterplot(x="waterfront present",y="Price",data=data,ax=axes[2,0])
sns.scatterplot(x="number of views",y="Price",data=data,ax=axes[2,1])
```

### Categorical/Categorical

## Multivariate

```python
multi=data.iloc[:,[7,8,9,10,20,22]]
multi.head()
```

```python
sns.pairplot(multi,diag_kind="kde")
```

```python
sns.pairplot(multi,hue="Number of schools nearby")
```

```python
sns.pairplot(multi,hue="number of views")
```

```python
sns.lmplot(data=multi,x="grade of the house",y="condition of the house",hue="Number of schools nearby",col="waterfront present")
```

```python
sns.heatmap(data=multi.corr(),annot=True,cmap="mako")
```

```python
sns.scatterplot(data=multi,x="grade of the house",y="Price",hue="waterfront present",palette="bright")
```
